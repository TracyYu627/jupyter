```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
np.random.randint(4)
```

```python
accounts = (712,713,714,715)
accountNameMap = {712:'Zhang San', 713:'Li Si', 714:'Wang Wu', 715:'Zhao Liu'}
products = ('CPU', 'Software', 'Maintenance')
priceMap = {'CPU': 30000, 'Software': 1000, 'Maintenance':500}
status = ('presented', 'pending', 'declined')
data = []
for i in range(100):
    randAccount = accounts[np.random.randint(4)]
    accountName = accountNameMap[randAccount]
    randProduct = products[np.random.randint(3)]
    price = priceMap[randProduct] + np.random.randint(100)
    randStatus = status[np.random.randint(3)]
    data.append((randAccount, accountName, randProduct,price,randStatus))

len(data)
```

```python
df = DataFrame(
    data,
    columns=['account','accountName','products','price','status']
)
```

```python
df.head()
```

```python
pd.pivot_table(df,index=['accountName'])
```

```python
pd.pivot_table(df,index=['products','status'],values=['price'], aggfunc='sum')
```

```python
pd.pivot_table(df,index=['products', 'accountName'],values=['price'], columns=['status'],aggfunc='max')
```
